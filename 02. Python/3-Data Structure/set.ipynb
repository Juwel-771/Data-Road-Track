```python
## creating set
my_set={1,2,3,4}
print(my_set)
```

```python
print(type(my_set))
```

```python
my_set.add(5)
print(my_set)
```

```python
my_set.discard(6)
print(my_set)
```

```python
remove_pop = my_set.pop()
print(remove_pop)
print(my_set)
```

```python
my_set.clear()
print(my_set)
```

```python
my_set = {1,2,3,4,5}
print(3 in my_set)
print(8 in my_set)
```

```python
set1={1,2,3,4,5}
set2={6,7,8,3,2}

union_set = set1.union(set2)
print(union_set)
```

```python
intersection_set = set1.intersection(set2)
print(intersection_set)
```

```python
print(set1.difference(set2))
```

```python
set2.difference(set1)
```

```python
my_set1 = {1,2,3}
my_set2 = {4,5,6}

print(my_set1.issubset(my_set2))
print(my_set1.issuperset(my_set2))
```
