```python
todo_list=["Buy Groceries", "Clean the house", "Pay Bills"]

## Adding tast
todo_list.append("Schedule a meeting")
todo_list.append("Go for a fun")

## Removing completed task
todo_list.remove("Clean the house")

## Checking if list remaining
if "Pay bills" in todo_list:
    print("Don't forget to pay bills")

print("To Do List remaining")
for task in todo_list:
    print(f"-{task}")
```

```python
# Organizing student grades
grades = [85, 92, 78, 90, 88]

# Adding a new grade
grades.append(95)

# Calculating the average grade
average_grade = sum(grades) / len(grades)
print(f"Average Grade: {average_grade:.2f}")

# Finding the highest and lowest grades
highest_grade = max(grades)
lowest_grade = min(grades)
print(f"Highest Grade: {highest_grade}")
print(f"Lowest Grade: {lowest_grade}")
```

```python
# Managing an inventory
inventory = ["apples", "bananas", "oranges", "grapes"]

# Adding a new item
inventory.append("strawberries")

# Removing an item that is out of stock
inventory.remove("bananas")

# Checking if an item is in stock
item = "oranges"
if item in inventory:
    print(f"{item} are in stock.")
else:
    print(f"{item} are out of stock.")

# Printing the inventory
print("Inventory List:")
for item in inventory:
    print(f"- {item}")
```
