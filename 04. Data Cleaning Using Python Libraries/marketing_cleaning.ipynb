```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('marketing_campaign_data_messy.csv')
print(f"Loaded Data: {df.shape[0]}row, {df.shape[1]} columns")
```

```python
print(df.columns.tolist())
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '-')
```

```python
print(df.columns.tolist())
```

```python
df
```

```python
spend_mask = df['spend'].astype(str).str.contains(r'\$')
print(df.loc[spend_mask, ['campaign_id', 'spend']].head(3))

df['spend'] = df['spend'].astype(str).str.replace(r'[^\d.-]','', regex=True)
df['spend'] = pd.to_numeric(df['spend'], errors='coerce')

print("FIX APPLIED")    
print(df.loc[spend_mask, ['campaign_id', 'spend']].head(3))
```
