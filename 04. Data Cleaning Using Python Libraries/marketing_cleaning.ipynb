```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('marketing_campaign_data_messy.csv')
print(f"Loaded Data: {df.shape[0]}row, {df.shape[1]} columns")
```

```python
print(df.columns.tolist())
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '-')
```

```python
print(df.columns.tolist())
```

```python
df
```

```python
spend_mask = df['spend'].astype(str).str.contains(r'\$')
print(df.loc[spend_mask, ['campaign_id', 'spend']].head(3))

df['spend'] = df['spend'].astype(str).str.replace(r'[^\d.-]','', regex=True)
df['spend'] = pd.to_numeric(df['spend'], errors='coerce')

print("FIX APPLIED")    
print(df.loc[spend_mask, ['campaign_id', 'spend']].head(3))
```

```python
print(df['channel'].unique())
```

```python
clean_up = {
    'Facebok':'Facebook',
    'Tik_Tok':'TikTok',
    'E-mail' :'Email',
    'Insta_gram' : 'Instagram',
    'Gogle' : 'Google Ads',
    'N/A' : np.nan
}

df['channel'] = df['channel'].replace(clean_up)

df['channel'].unique()
```

```python
df
```

```python
df['active'].unique()
```

```python
clean_active = {
    'Y' : True,
    'Yes' : True,
    '1' : True,
    1 : True,
    'True' : True,
    'N' : False,
    'No' : False,
    '0' : False,
    0 : False,
    'False' : False
}

df['active'] = df['active'].map(clean_active).fillna(False).astype(bool)
print(df['active'].unique())
```

```python
df
```

```python
df['start_date'] = pd.to_datetime(df['start_date'], errors='coerce')
df['end_date'] = pd.to_datetime(df['end_date'], dayfirst=False, errors='coerce')

print(df['start_date'].dtype)
```

```python
df
```

```python
df = df.loc[:, ~df.columns.duplicated()]
```

```python
df
```

```python
impossible_mask = df['clicks'] > df['impressions']
print(df.loc[impossible_mask, ['campaign_id', 'impressions', 'clicks']].head(3))
```

```python
travel_mask=df['end_date'] < df['start_date']
print(df.loc[travel_mask, ['campaign_id', 'start_date', 'end_date']].head(3))

df.loc[travel_mask, 'end_date'] = df.loc[travel_mask, 'start_date']+pd.Timedelta(days=30)
print(df.loc[travel_mask, ['campaign_id', 'start_date', 'end_date']].head(3))
```

```python
Q1 = df['spend'].quantile(0.25)
Q3 = df['spend'].quantile(0.75)

IQR = Q3-Q1
upper_limit=Q3+(3*IQR)

outlier_mask = df['spend'] > upper_limit
print(df.loc[outlier_mask, ['campaign_id', 'spend']].head(3))

df.loc[outlier_mask, 'spend'] = upper_limit
print(df.loc[outlier_mask, ['campaign_id', 'spend']].head(3))
```

```python
df['season'] = df['campaign_name'].str.extract(r'Q\d_([^_]+)_')

print(df[['campaign_name','season']].head(3))
```
