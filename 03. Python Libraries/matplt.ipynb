```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
## Basic plot
x = np.array([2023, 2024, 2025, 2026])
y = np.array([15, 25, 30, 20])

plt.plot(x,y, marker='o')
plt.show()
```

```python
## customize plot
x = np.array([2023, 2024, 2025, 2026])
y = np.array([10, 20, 15, 25])

plt.plot(x,y, marker='o',   markersize=10, 
                            markerfacecolor='red', 
                            markeredgecolor='blue',
                            linestyle='dotted')
plt.show()
```

```python
x = np.array([2023, 2024, 2025, 2026])
y1 = np.array([10, 20, 15, 25])
y2 = np.array([15, 28, 18, 23])
y3 = np.array([18, 21, 25, 20])

plt.title("Class Size")
plt.xlabel("Class")
plt.ylabel("Student")

plt.grid()
# plt.grid(axis='x')
plt.grid(axis='y')

plt.plot(x,y1, marker="o")
plt.plot(x,y2, marker="o")
plt.plot(x,y3, marker="o")
plt.xticks(x)
```

```python
## Bar chart
categories = np.array(['Grains', 'Fruits', 'Vegetable', 'Protein', 'Meat', 'Liquid'])
values = np.array([4,3,2,5,3,1])

plt.bar(categories, values, color="green")
plt.title("Daily consumption")
plt.xlabel("Food")
plt.ylabel("Quantity")
plt.show()
```

```python
## pie chart
categories=np.array(['Freshman', 'Sophomorse', 'Juniors', 'Senior'])
value = np.array([300, 250, 275, 225])

plt.pie(value, labels=categories, 
                autopct="%1.1f%%",
                explode=[0,0,0,0.2],
                shadow=True
                )
```

```python
## scatter chart
x = np.array([0,1,1,2,3,4,5,6,7,7,8])
y = np.array([55,60,65,62,68,70,75,78,82,85,87])

x2 = np.array([0,1,1,2,3,4,5,6,7,7,8])
y2 = np.array([50,55,60,67,74,75,80,85,87,75,80])

plt.title("Test Score")
plt.xlabel("Hours Studied")
plt.ylabel("Grade")

plt.scatter(x,y,    color='red',
                    alpha =0.5,
                    s=100,
                    label="Class A")

plt.scatter(x2, y2, color="blue",
                    alpha=0.5,
                    s=100,
                    label="Class B")

plt.legend()
plt.show()
```

```python
## histogram
score = np.random.normal(loc=80, scale=50, size=100)
plt.hist(score, bins=15, color="lightgreen", edgecolor="black")

plt.title("Exam Score")
plt.xlabel("Score")
plt.ylabel("Student")

plt.show()
```
