```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

## creating line plot
plt.plot(x,y)
plt.xlabel('X Axis')
plt.ylabel('Y Axix')
plt.title("Basic Plot")
plt.show()
```

```python
## customize line plot
x=[1,2,3,4,5]
y=[1,4,9,16,25]

plt.plot(x,y, color='red', linestyle='-', marker='o', linewidth='2', markersize='5')
plt.grid(True)
```

```python
## multiple plot
x = [1,2,3,4]
y1 = [1,5,8,14]
y2 = [1,3,5,7]

plt.subplot(2,2,1)
plt.plot(x,y1, color='red')
plt.title("Plot 1")

plt.subplot(2,2,2)
plt.plot(x,y2, color='green')
plt.title("Plot 2")

plt.subplot(2,2,3)
plt.plot(y1,y2, color='blue')
plt.title("Plot 3")

plt.subplot(2,2,4)
plt.plot(y2,y1, color='black')
plt.title("Plot 4")
```
