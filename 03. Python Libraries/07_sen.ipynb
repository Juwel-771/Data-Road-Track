```python
import seaborn as sns
import pandas as pd
```

```python
sns.get_dataset_names()
```

```python
pen = sns.load_dataset('penguins')
```

```python
pen['species'].value_counts()
```

```python
pen['island'].value_counts()
```

```python
## scatter plot
pen
```

```python
sns.set_style('darkgrid')
sns.set_context("paper")
sns.scatterplot(x='flipper_length_mm', y='body_mass_g', data=pen, hue='island')
sns.despine(left=True)
```

```python
sns.scatterplot(x='species', y='body_mass_g', data=pen, hue='island', alpha=0.5)
```

```python
# strip plot
sns.stripplot(x='species', y='body_mass_g', data=pen, hue='sex', dodge=True, jitter=True)
```

```python
#swarm plot
sns.swarmplot(x='species', y='body_mass_g', data=pen, hue='island')
```

```python
#Histogram 
sns.histplot(x='body_mass_g', data=pen, hue='sex', multiple='stack')
```

```python
# regression plot
sns.regplot(x='body_mass_g', y='flipper_length_mm', data=pen, color='red')
```

```python
# line plot
sns.lineplot(x='body_mass_g', y='flipper_length_mm', data=pen, hue='island', style='sex')
```

```python
## joint plot
sns.jointplot(x='body_mass_g', y='flipper_length_mm', data=pen, hue='island', kind='scatter')
```

```python
##Bar plot
pen
```

```python
sns.barplot(x='species', y='body_mass_g', hue='sex', data=pen)
```

```python
## count plot
sns.countplot(data=pen, x='species')
```

```python
## box plot
sns.boxplot(x='species', y='body_mass_g', data=pen, hue='sex')
```

```python
## violin plot
sns.violinplot(x='species', y='body_mass_g', data=pen, hue='sex')
```

```python
#kde plot
sns.kdeplot(data=pen, x='body_mass_g', hue='species', fill=False)
```
