```python
## custom index
sale_series = [0,5,2,4,1]
items = ['coffee', 'tea', 'cola', 'soda', 'sugar']

sale_series_custom_index=pd.Series(sale_series, index=items, name='Sales')
```

```python
sale_series_custom_index
```

```python
sale_series_custom_index.index.isin(['coffee', 'tea'])
```
